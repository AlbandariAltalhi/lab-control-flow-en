## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
upper_case_words = [word.upper() for word in words]
print(upper_case_words)
```

**Create a new list containing only words with 5 or more letters**

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story', 'island']
filtered_words = [word for word in words if len(word) >= 5]
print(filtered_words)
```

**Print the first word starting with "t"**

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story', 'island']
first_t_word = next((word for word in words if word.startswith('t')), None)
print(first_t_word)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares = [x**2 for x in range(1, 11)]
print(squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
odd_squares = [x**2 for x in range(1, 11) if x % 2 != 0]
print(odd_squares)
```

:**Create a list with the squares of all multiples of 8 below 1000**

```python
multiples_of_8_squares = [x**2 for x in range(8, 1000, 8)]
print(multiples_of_8_squares)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
len(people)
```

**How many people have kids**?

```python
people_with_kids = [person for person in people if person["n_kids"] > 0]
print(len(people_with_kids))
```

**How many kids do they have in total?**






```python
total_kids = sum(person["n_kids"] for person in people if person["n_kids"] > 0)
print(total_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
people_in_one_year = []
for person in people:
    if person["name"].endswith('a'):
        person["n_kids"] += 1
    people_in_one_year.append(person)

print(people_in_one_year)
```
